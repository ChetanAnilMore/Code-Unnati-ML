## Prepare data for machine learning : 

```python
# We know that the machine learning agorithms are mathematical expressioins so we need to provide the data to the  algorithm and specific format 
# Encoding 
```

```python
# Divide the data based on their data types 
df_cat = df_houseprice.select_dtypes(include=['object'])
df_cat.columns
```

```python
# use dummy encoding
df_var = pd.get_dummies(data = df_cat, drop_first=True)
df_var = df_var.reset_index()
```

```python
df_var2 = pd.get_dummies(data = df_cat)
```

```python
df_var
```

```python
df_var2
```

```python
df_num = df_houseprice.select_dtypes(include=[np.number])

df_num
```

```python
df_num.reset_index(inplace=True)
df_num
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
```

```python
scale = ss.fit_transform(df_num)
```

```python
scale 
```

```python
scale = pd.DataFrame(scale,columns = df_num.columns)
```

```python
scale = scale.reset_index()
scale
```

```python
df_merged = pd.concat([scale,df_var],axis=1)
```

```python
df_merged.drop('Sold',axis=1,inplace=True)
```

```python
df_merged
```

```python
# Seperate the target variable and independent variable
target = df_houseprice['Sold']
```

```python
# independent variable
X = df_merged
```

```python
# Split the data for training and testing 
import statsmodels 
import statsmodels.api as sm 

X = sm.add_constant(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,target,random_state=10,test_size=0.2)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train,y_train)
```

```python
predictions = tree_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
```

```python
accuracy
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_model = RandomForestClassifier(n_estimators=300)
```

```python
rfc_model.fit(X_train, y_train)
```

```python
predictions = rfc_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
```

```python
accuracy
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
```

```python
predictions = lr_model.predict(X_test)
accuracy = accuracy_score(y_test,predictions)
accuracy
```

## KNeighbours Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knc_model = KNeighborsClassifier()
knc_model.fit(X_train, y_train)
```

```python
predictions = knc_model.predict(X_test)
accuracy = accuracy_score(y_test,predictions)
accuracy
```

## Linear Discriminant Analysis 

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda_model = LinearDiscriminantAnalysis()
lda_model.fit(X_train,y_train)
```

```python
predictions = lda_model.predict(X_test)
accuracy = accuracy_score(y_test,predictions)
accuracy
```

## Quadratic Discriminant Analysis

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

qda_model = QuadraticDiscriminantAnalysis()
qda_model.fit(X_train, y_train)
```

```python
predictions = qda_model.predict(X_test)
accuracy = accuracy_score(y_test,predictions)
accuracy
```

## Artificial Neural Networks

```python
import tensorflow
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units=64,activation='relu',input_shape = (X_train.shape[1],)))
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10, batch_size = 32, validation_split = 0.2)
```

```python
predictions = model.predict(X_test)
predictions = (predictions > 0.5)
accuracy = accuracy_score(y_test, predictions)
accuracy
```
